```python
import os
import random
import torchmetrics
import torch
from PIL import Image
import torchvision.transforms as transforms
from tqdm import tqdm
```

```python
test_dir = '/mnt/g/visuals_from_server/512-64-700-test'
```

```python
indexes = range(2400)
group_size = 5
```

```python
groups = []
i=0
# create 2400 groups of 4 images
while i < 2400:
    g = [i]
    # randomly sample 3 other images
    for j in range(group_size-1):
        new_index = i + j + 1

        g.append(new_index)

    i += group_size
    groups.append(g)

print(len(groups))
```

```python
lpips_all_values = []
lpips_avg_gorups = []

for g in tqdm(groups):
    # calculate the lpips for each group
    lpips = torchmetrics.image.lpip.LearnedPerceptualImagePatchSimilarity(normalize=True).cuda()
    images = []
    lpips_values = []
    for i in g:
        image_path = os.path.join(test_dir, f'{i}.png')
        image = Image.open(image_path)

        images.append(transforms.ToTensor()(image).to('cuda'))
        
    images = torch.stack(images)

    # do full pairwise comparison
    for i in range(group_size):
        for j in range(i+1, group_size):
            lpips_values.append(lpips(images[i].unsqueeze(0), images[j].unsqueeze(0)).item())

    #print(lpips_values)

    lpips_avg_gorups.append(sum(lpips_values) / len(lpips_values))
    lpips_all_values.extend(lpips_values)

#print(lpips_avg_gorups)


        
```

```python
print(lpips_avg_gorups)

print(sum(lpips_avg_gorups) / len(lpips_avg_gorups))

# calcualte the std over all values
import statistics
print(statistics.stdev(lpips_all_values))
```
