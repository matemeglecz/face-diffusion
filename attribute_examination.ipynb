```python
import pandas as pd
```

```python
# load data/CelebAMask-HQ/CelebA-HQ-attribute-anno.txt
df = pd.read_csv('data/CelebAMask-HQ/CelebAMask-HQ-attribute-anno.txt', sep='\s+', header=1)
```

```python
df.head()
```

```python
# replace -1 with 0
df = df.replace(-1, 0)
```

```python
# create correlation matrix of attributes
corr = df.corr()
corr
```

```python
# filter the matrix for attribute list
attribute_list = ['Male', 'Young', 'Bald', 'Bangs', 'Receding_Hairline', 'Black_Hair', 'Blond_Hair', 'Brown_Hair', 'Gray_Hair', 'Straight_Hair', 'Wavy_Hair', 'No_Beard', 'Goatee', 'Mustache', 'Sideburns', 'Narrow_Eyes', 'Oval_Face', 'Pale_Skin', 'Pointy_Nose']

corr = corr[attribute_list]
corr = corr.loc[attribute_list]
corr
```

```python
# visualize correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 20))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.savefig('correlation_matrix.png')
plt.show()
```
